### 1. Analyze the relationship between Loan and Borrower's gender, age, and degree

#### 1.1 Relationship between Loan and Borrower's gender

```python
data.info()
```

#### 1.2.1 Relationship between Loan and Borrowers's age

```python
# Calculate the total loan amount by age
df_age = data.groupby('Age')['Loan_Amount'].sum()
```

```python
df_age = pd.DataFrame(df_age)
df_age.head()
```

```python
# Calculate the cummulative of loan amount
df_age['Cum_Amount'] = df_age['Loan_Amount'].cumsum()
```

```python
df_age.head()
```

```python
# Calculate the ratio of cumulative loan amount
df_age['Cum_Amount_Ratio'] = df_age['Cum_Amount'] / df_age['Loan_Amount'].sum()
```

```python
df_age.head()
```

```python
# Obtain the age group that covers 80% of the Loans using The Pareto Principle (80-20 Rule)
age_group = df_age[df_age['Cum_Amount_Ratio'] > 0.8].index
```

```python
print(f'80% of the Loans lend to the age groups between 18 and {age_group[0]} years old.')
```

```python
# Calculate cumulative ratio
cum_ratio = df_age.loc[age_group[0], 'Cum_Amount_Ratio']
```

```python
cum_ratio
```

```python
# Visualize the result in one figure
plt.figure(figsize = (16,9))
plt.bar(x = df_age.index, 
        height = df_age['Loan_Amount'],
        color = 'blueviolet',
        alpha = 0.5,
        width = 0.8)
plt.xlabel('Age', fontsize = 15, labelpad = 20)
plt.ylabel('Loan Amount', fontsize = 15, labelpad = 20)

# Add a guideline that parallel to y axis
plt.axvline(x = age_group[0], color = 'brown', linestyle = '--', alpha = 0.8)

df_age['Cum_Amount_Ratio'].plot(style = '--ob', 
                             secondary_y = True)
plt.ylabel('Cummulative Amount Ratio', fontsize = 15, labelpad = 20)

plt.text(age_group[0] + 1, 
         cum_ratio, 
         'Cummulative Percentage: %.2f%%' % (cum_ratio*100), 
         color = 'green',
         fontsize = 15)

plt.show()
```

#### 1.4 Use def functioin to analyze the relationship between Loan and Borrower's Phone, Household, Video, Credit and App verification

```python
def trans(data, col, ind):
    df = pd.pivot_table(data = data, columns = col, index = ind,
                        values = 'ListingId', aggfunc = np.size)
    df['No_of_Loans'] = df.apply(np.sum, axis = 1)
    df['Loan_Percentage'] = df['No_of_Loans'] / df['No_of_Loans'].sum()
    df['Pastdue_Percentage'] = df['Past due'] / df['No_of_Loans']
    plt.figure(figsize = (16,12))
    plt.subplot(121)
    plt.pie(x = df['Loan_Percentage'], labels = ['Verified','Not Verified'],
            colors = ['lightblue', 'pink'], autopct = '%.2f%%',
            pctdistance = 0.5, labeldistance = 1.1)
    plt.title('Number of Loan Application Percentage by %s' % ind, size = 15)
    plt.subplot(122)
    plt.bar(x = df.index, height = df['Pastdue_Percentage'], color = ['lightblue','pink'])
    plt.xticks(df.index,
               labels = ['Not Verified', 'Verified'])
    plt.title('Past Due Percentage by %s' % ind, size = 15, pad = 20)
    plt.suptitle("Borrower's Persona by %s" % ind, size = 20)
    plt.show()
    return df
```

```python
trans(data, 'Status', 'Phone_Verify')
```

```python
trans(data, 'Status', 'Household_Verify')
```

```python
trans(data, 'Status', 'Credit_Verify')
```

```python
trans(data, 'Status', 'Video_Verify')
```

```python
trans(data, 'Status', 'App_Verify')
```
